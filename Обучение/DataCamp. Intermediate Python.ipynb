### Matplotlib

```python
# Импортируем библиотеку для визуализации данных и загрузим датасет
import matplotlib.pyplot as plt
year = [1950, 1970, 1990, 2010]
pop = [2.519, 3.692, 5.263, 6.972]
```

#### Linear diagramm
<st>
matplotlib.pyplot.plot(*args, scalex=True, scaley=True, data=None, **kwargs)

```python
# Функция plot задает вводные для строительства линейчатой диаграммы (горизонталь, вертикаль)
plt.plot(year, pop)
# Функция show строит график (визуализацию)
plt.show()
```

#### Scatter diagramm
<st>
matplotlib.pyplot.scatter(x, y, s=None, c=None, marker=None, cmap=None, norm=None, vmin=None, vmax=None, alpha=None, linewidths=None, *, edgecolors=None, plotnonfinite=False, data=None, **kwargs)


```python
# Функция scatter задает вводные для точечной диаграммы
plt.scatter(year, pop) # Если еще задать s (plt.scatter(year, pop, s = variable), то толщина тточек может также быть информацией, 
# c = словарь с цветами, alpha = прозрачность (от 0 до 1)
plt.show()
```

Оси на всех графиках, использующих Matplotlib, по умолчанию линейны, yscale() и xscale() методы библиотеки matplotlib.pyplot можно использовать для изменения масштаба по оси y или оси x на логарифмический соответственно.

xscale('log')
<br>
Можно использовать три разных масштаба:
<br>

'linear' – линейный масштаб (используется по умолчанию);
<br>
'log' – логарифмический масштаб;
<br>
'symlog' – вблизи нуля (в указанных пределах) масштаб линейный, а в остальной области – логарифмический.
<br>

Логарифмический масштаб формируется по основанию 10. Если нужно изменить и указать другое основание, то это делается с помощью параметра base:
<br>
ax.set_yscale('log', base=5)
<br>

Построение графиков и пользование ими становится затруднительными, если величины, откладываемые вдоль координатных осей, изменяются очень в больших пределах. В этом случае используется логарифмический масштаб, позволяющий значительно расширить пределы изменения функций наносимых на график, без увеличения размеров чертежа. Для этого вдоль координатных осей откладывается вместо значений функций десятичные логарифмы этих значений, а полученным точкам присваиваются названия откладываемых значений. В результате применения логарифмического масштаба по одной из координатных осей кривизна графиков уменьшается, а зависимости, близкие к показательной функции, приближаются к прямым линиям
<br>
Если по одной оси координат принят логарифмический масштаб, а по другой – обычный (линейный), то такую координатную сетку называют полулогарифмической

#### Customization

```python
import matplotlib.pyplot as plt
year = [1950, 1951, 1952, 2100] 
pop = [2.538, 2.57, 2.62, 10.85]

# Add more data
year = [1800, 1850, 1900] + year
pop = [1.0, 1.262, 1.650] + pop

plt.plot(year, pop)
plt.grid(True) # Добавляет черты (сетку)
plt.text(1950, 2.519, "Подпись значения") # Можем подписать некоторые значения, точки
plt.xlabel('Year') # Добавляем подпись по оси х
plt.ylabel('Population') # Добавляем подпись по оси у
plt.title('World Population Projections') # Добавляем подпись диаграммы
plt.yticks([0, 2, 4, 6, 8, 10], # Добавляем метки по оси у
           ['0', '2B', '4B', '6B', '8B', '10B']) # Добавляем подписи для меток по оси у
plt.show()
```

### Dictionaries

```python
# Создаем словарь: "afghanistan":30.55, где "afghanistan" - ключ, 30.55 - значение
world = {"afghanistan":30.55, "albania":2.77, "algeria":39.21}
# Ищем значение по ключу
world["albania"]
```

```python
# Можно вывести список ключей 
print(world.keys())
```

```python
# Добавляем новое значение в справочник
world['italy'] = 5.36
print(world)
```

```python
# Проверяем наличие ключа в справочнике
print('albania' in world)
```

```python
# Изменять значения можно переписывая предыдущие
world['italy'] = 8.36
print(world)
```

```python
# Удаляем значения из словаря
del(world['italy'])
print(world)
```

#### Словарь со словарями

```python
# Dictionary of dictionaries
europe = { 'spain': { 'capital':'madrid', 'population':46.77 },
           'france': { 'capital':'paris', 'population':66.03 },
           'germany': { 'capital':'berlin', 'population':80.62 },
           'norway': { 'capital':'oslo', 'population':5.084 } }


# Print out the capital of France
print(europe['france']['capital'])

# Create sub-dictionary data
data = {'capital':'rome', 'population':59.83}

# Add data to europe under key 'italy'
europe['italy'] = data
print(europe)
```

### Pandas

DataFrame - название таблицы в pandas

Строки содержат значения, столбцы содержат переменные
<br>
Каждая строка имеет уникальную метку строки. Столбцы и переменные также имеют метки

```python
# Создаем DataFrame через справочник
dict = { "country":["Brazil", "Russia", "India", "China", "South Africa"],
        "capital":["Brasilia", "Moscow", "New Delhi", "Beijing", "Pretoria"],
        "area":[8.516, 17.10, 3.286, 9.597, 1.221],
        "population":[200.4, 143.5, 1252, 1357, 52.98] }

# keys (column labels)
# values (data, column by column)

import pandas as pd
brics = pd.DataFrame(dict)
print(brics)
```

```python
# Можно добавить индексы строк
brics.index = ["BR", "RU", "IN", "CH", "SA"]
print(brics)
```
